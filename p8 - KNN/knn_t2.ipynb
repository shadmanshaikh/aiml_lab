```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
from sklearn import datasets
print("loading the dataset")
iris = datasets.load_iris()

x_test , x_train , y_test , y_train = train_test_split(iris.data, iris.target , test_size =0.1)
print("the test data is splitted : " ,x_test.shape , y_test.shape)
print("the train data is splitted : " ,x_train.shape , y_train.shape)

print("the target names are : ")
for i in range(len(iris.target_names)):
    print("lable" + str(i) + "-" + str(iris.target_names[i]))

classifier = KNeighborsClassifier(n_neighbors=1)

classifier.fit(x_train , y_train)
y_pred = classifier.predict(x_test)

print("the result using KNN where n =1 ")
for i in range(0 , len(x_test)):
    print("sample: " + str(x_test[i]) + "actual : " + str(y_test[i]) + "predict :" + str(y_pred[i]))

print("the accuracy of the model is : " , classifier.score(x_test, y_test))

from sklearn.metrics import confusion_matrix , classification_report

print("the confusion matrix is :" , confusion_matrix(y_test, y_pred))
print("the classification report is : " , classification_report(y_test , y_pred))

 



```
